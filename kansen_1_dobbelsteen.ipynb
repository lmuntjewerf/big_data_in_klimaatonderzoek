```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

### Gooien met 1 dobbelsteen

```python
def gooi_met_een_dobbelsteen(hoevaak, kanten_van_dobbelsteen= [1, 2, 3, 4, 5, 6]):
    np.random.seed(2020)
    dobbelsteen = []
    for _ in range(hoevaak):
        rol = np.random.choice(kanten_van_dobbelsteen)
        dobbelsteen.append(rol)

    return dobbelsteen
```

```python
een_dobbelsteen_weinig = gooi_met_een_dobbelsteen(30)
een_dobbelsteen_veel = gooi_met_een_dobbelsteen(30000)
```

```python
fig, (ax1,ax2) = plt.subplots(ncols=2, sharey=False, figsize=(12,4))
sns.countplot(x=een_dobbelsteen_weinig, ax=ax1)
sns.countplot(x=een_dobbelsteen_veel, ax=ax2)
plt.tight_layout()
plt.savefig('figuren/een_dobbelsteen.png', dpi=200, )
# plt.show()
```

### Gooien met twee dobbelstenen

```python
def gooi_met_twee_dobbelstenen(hoevaak, kanten_van_dobbelsteen= [1, 2, 3, 4, 5, 6]):
    np.random.seed(2020)
    twee_dobbelstenen = []
    for _ in range(hoevaak):
        rol1 = np.random.choice(kanten_van_dobbelsteen)
        rol2 = np.random.choice(kanten_van_dobbelsteen)
        optelsom = rol1+rol2
        twee_dobbelstenen.append(optelsom)

    return twee_dobbelstenen
```

```python
twee_dobbelstenen_weinig = gooi_met_twee_dobbelstenen(50)
twee_dobbelstenen_veel = gooi_met_twee_dobbelstenen(50000)


fig, (ax1,ax2) = plt.subplots(ncols=2, sharey=False, figsize=(12,4))
sns.countplot(x=twee_dobbelstenen_weinig, ax=ax1)
sns.countplot(x=twee_dobbelstenen_veel, ax=ax2)

plt.tight_layout()
plt.savefig('figuren/twee_dobbelstenen.png', dpi=200, )
plt.show()
```
